```python
df.drop('Location',axis = 1, inplace= True)
```

```python
# getting dummies for Fuel_Type	Transmission Owner_Type
temp_df = pd.get_dummies(df, columns= ['Name','Fuel_Type','Transmission','Owner_Type'])
temp_df
```

```python
temp_df.replace(False, 0, inplace = True)
temp_df.replace(True, 1, inplace = True)
```

```python
temp_df
```

```python
X = temp_df.drop(['Price'], axis = 1)
```

```python
y = temp_df[['Price']]
```

```python
X
```

```python
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y , random_state=10, test_size= 0.2)
```

```python
len(xtrain),len(xtest), len(ytrain),len(ytest)
```

```python
rfc = RandomForestRegressor(n_estimators=200)
```

```python
rfc.fit(xtrain, ytrain)
```

```python
ypred = rfc.predict(xtest)
ypred
```

```python
r2_score(ytest, ypred)
```
